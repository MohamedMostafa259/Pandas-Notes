# Selecting Data with `.query()` Method
The `query()` method accepts an input string that it will use to select rows to return from the table.

**N.B.** `.query()` is less efficient than `[]`, `.loc[]`, and `.iloc[]`

```python
# single condition
sales.query("weekly_sales > 2e4")
```

```python
# multiple conditions 
sales.query("weekly_sales > 2e4 and weekly_sales < 5e4")
```

```python
sales['date'] = pd.to_datetime(sales['date'])
sales.query("(type=='A' or is_holiday==True) and date.dt.year==2010")
```
