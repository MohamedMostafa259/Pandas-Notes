```python
import pandas as pd
```

```python
sales = pd.read_csv("https://raw.githubusercontent.com/MohamedMostafa259/Pandas-Notes/refs/heads/main/Data/sales_sample.csv")
sales # last 3 rows are the same
```

# Setting a Column as The Index

```python
print(sales)
sales.set_index("type", inplace=True)
sales
```

# Resetting the Index column (Contains Row Labels)

```python
# Returns the Index column to its original place in the DataFrame
sales.reset_index(inplace=True)
sales
```

# Dropping The Index Column

```python
sales.set_index("type", inplace=True)
print(sales)

# drops the Index column (removes it and doesn't return it again as a normal column in the DataFrame)
sales.reset_index(drop=True, inplace=True)
sales
```

# Multi-level Index a.k.a. Hierarchical Index

```python
sales.set_index(["store", "department"], inplace=True)
sales
```

```python
# subset the outer level with a list → [outer, outer, ...]
sales.loc[[1, 7]]
```

```python
# subset the inner levels with a list of tuples → [(outer, inner), (outer, inner), ...]
sales.loc[[(1, 1), (7, 3)]]
```

# Sorting By Index Values

```python
# By default: sorts ascendingly by the 1st index col then by the 2nd then ...
sales.sort_index()
```

```python
# Controlling sort_index
sales.sort_index(level=["department", "store"], ascending=[True, False])
```

```python
sales = pd.read_csv("https://raw.githubusercontent.com/MohamedMostafa259/Pandas-Notes/refs/heads/main/Data/sales_sample.csv") 
# sales.set_index("type", inplace=True) --equals-- sales = pd.read_csv("sales_sample.csv", index_col=1)
sales
```

```python
sales_sorted = sales.set_index(["type", "department"]).sort_index()
sales_sorted
```

```python
# Slicing the outer index level
sales_sorted.loc["A":"B"] # Note that the final value "B" is included (.loc[] is inclusive, but .iloc[] is exclusive)
```

```python
# Slicing the inner index level → pass the 1st and 2nd indexes as tuples → start_index:(outer, inner) : end_index(outer, inner)
sales_sorted.loc[('A', 1):('C', 2)]
```

```python
# slicing on columns 
print(sales_sorted.loc[:, "weekly_sales":"temperature_c"])
# slicing on rows and cols 
sales_sorted.loc[('A', 1):('C', 2), "weekly_sales":"temperature_c"]
```

```python
sales_sorted = sales.set_index("date").sort_index()
sales_sorted
```

```python
# You don't need to specify the full date
print(sales_sorted.loc["2010":"2012"])
sales_sorted.loc["2010-03":"2010-06"] # "2010-06-04" isn't included because it's greater than the end index, "2010-06-00" 
```
