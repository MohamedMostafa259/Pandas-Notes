# Visualizing DataFrames

```python
import matplotlib.pyplot as plt

"""
Explanation of each column of the avocados dataset:

Date: The date when the data was recorded.

AveragePrice: The average price of a single avocado.

Total Volume: The total number of avocados sold.

4046: The number of avocados sold with PLU (Price Look-Up) code 4046, which typically refers to small or extra-small avocados.

4225: The number of avocados sold with PLU code 4225, which usually refers to medium avocados.

4770: The number of avocados sold with PLU code 4770, often referring to large or extra-large avocados.

Total Bags: The total number of bags sold.

Small Bags: The number of small bags of avocados sold.

Large Bags: The number of large bags of avocados sold.

XLarge Bags: The number of extra-large bags of avocados sold.

type: The type of avocado sale, which can be "conventional" or "organic".

year: The year when the data was recorded.

region: The geographical region where the data was recorded (e.g., a specific city or area).
"""

avocados = pd.read_csv("https://raw.githubusercontent.com/MohamedMostafa259/Pandas-Notes/refs/heads/main/Data/avocados.csv")
avocados.head()
```

#### Histograms
Increasing or decreasing the number of bins can give us an idea of what the distribution looks like

```python
avocados["AveragePrice"].hist(bins=20)
plt.show()
```

#### Bar Plots
It can show the relationship between a categorical variable and a numerical variable

```python
avg_price_by_type = avocados.groupby("type")["AveragePrice"].mean()
print(avg_price_by_type)

# rot=45 → rotates the x-axis labels by 45 degrees to make them easier to read
avg_price_by_type.plot(kind="bar", title="Average Price By Type", rot=45)
plt.show()
```

#### Line Plots
It is great for visualizing changes in numerical variables over time

```python
avocados["Date"] = pd.to_datetime(avocados["Date"])
avocados.sort_values("Date", inplace=True)
avocados.plot(x="Date", y="AveragePrice", kind="line", rot=45, figsize=(10, 8))
plt.show()
```

#### Scatter Plots
They are great for visualizing relationships between two numeric variables

***IMPORTANT:***
If your x-axis data ranges from very small to very large values, a logarithmic scale can make the plot easier to read and interpret by compressing the wide range into a more manageable scale.

```python
ax = avocados.plot(x="Total Volume", y="AveragePrice", kind="scatter", title="Total Number of Avocados Sold vs. Average Price")
ax.set_xscale("log")
plt.show()
```

### Layering Plots

```python
# alpha arg is useful when two charts cross each other
avocados[avocados["type"] == "conventional"]["AveragePrice"].hist(alpha=0.7, bins=20)
avocados[avocados["type"] == "organic"]["AveragePrice"].hist(alpha=0.7, bins=20)

plt.legend(["conventional", "organic"])
plt.show()
```
